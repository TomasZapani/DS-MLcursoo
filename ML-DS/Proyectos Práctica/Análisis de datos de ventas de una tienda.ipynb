```python
#Calc del promedio de los ingresos mensuales
meses = df['Fecha'].dt.month.to_numpy()#Ac谩 extraje de la serie fecha la parte de meses y lo conv en array
ingresos = df['Total Venta'].to_numpy()#Aca extraje de la serie total venta y lo conv en array 

meses_unicos = np.unique(meses)#Ac谩 encontramos los meses 煤nicos sin repetir, es decir si en el array hay:enero,enero,marzo,abril,abril, solo tomar谩 los meses 煤nicos sin repetir

for mes in meses_unicos:#que mes recorra meses unicos 
    ingresos_total = np.mean(ingresos[meses == mes])#Ac谩 calc el promedio de ingresos de todos los meses, filtramos los ingresos de todos los meses 
    print(f"Mes {mes}: Ingreso Promedio ${ingresos_total:.2f}")
```

```python
#Calc la desv est谩ndar de los ingresos mensuales
meses = df['Fecha'].dt.month.to_numpy()#Ac谩 extraje de la serie fecha la parte de meses y lo conv en array
ingresos = df['Total Venta'].to_numpy()#Aca extraje de la serie total venta y lo conv en array 

meses_unicos = np.unique(meses)#Ac谩 encontramos los meses 煤nicos sin repetir, es decir si en el array hay:enero,enero,marzo,abril,abril, solo tomar谩 los meses 煤nicos sin repetir

for mes in meses_unicos:#que mes recorra meses unicos 
    ingresos_total = np.std(ingresos[meses == mes]) 
    print(f"Mes {mes}: Desv est谩ndar ${ingresos_total:.2f}")
```

```python
#Total de ventas por prod
import numpy as np

# Convertir columnas a NumPy
total_ventas = df['Total Venta'].to_numpy()
productos = df['Producto'].to_numpy()

# Obtener productos 煤nicos y calcular ventas totales
prod_unicos = np.unique(productos)
ventas_por_producto = [np.sum(total_ventas[productos == prod]) for prod in prod_unicos]

# Combinar productos y ventas en una lista de tuplas y ordenarlas de mayor a menor
productos_ordenados = sorted(zip(prod_unicos, ventas_por_producto), key=lambda x: x[1], reverse=True)

# Imprimir resultados ordenados
for prod, total in productos_ordenados:
    print(f"Producto: {prod} - Total Ventas: {total}")
```

```python
#Ganancias max por prod
import numpy as np

# Convertir columnas a NumPy
total_ventas = df['Total Venta'].to_numpy()
productos = df['Producto'].to_numpy()

# Obtener productos 煤nicos y calcular ventas totales
prod_unicos = np.unique(productos)
ventas_por_producto = [np.max(total_ventas[productos == prod]) for prod in prod_unicos]

# Combinar productos y ventas en una lista de tuplas y ordenarlas de mayor a menor
productos_ordenados = sorted(zip(prod_unicos, ventas_por_producto), key=lambda x: x[1], reverse=True)

# Imprimir resultados ordenados
for prod, total in productos_ordenados:
    print(f"Producto: {prod} - Ganancias m谩x: {total}")
```
