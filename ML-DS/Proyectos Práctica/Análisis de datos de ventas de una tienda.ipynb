Obtener y limpiar los datos

    Te puedo pasar un dataset de ventas en formato CSV o podés generar uno aleatorio.
    Usa Pandas para cargar el dataset y revisar datos faltantes.
    Elimina valores nulos y transforma formatos de fecha y precios.

```python
df = pd.read_csv('datos_ventas.csv')
df
```

```python
pd.isna(df).sum()
```

```python
type(df['Fecha'][0])
```

```python
type(df['Precio Unitario'][0])
```

```python
df['Fecha'] = pd.to_datetime(df['Fecha'])
type(df['Fecha'][0])
```

```python
df['Precio Unitario'] = df['Precio Unitario'].astype(float)
type(df['Precio Unitario'][0])
```

Análisis exploratorio con Pandas

    Filtra los datos por mes, categoría de producto o cliente.
    Agrupa por categoría o por mes para calcular ingresos totales.
    Identifica los productos más vendidos.

```python
#Flitracion de ventas por el mes de Julio
df['Mes'] = df['Fecha'].dt.month
mes_julio = df[df['Mes'] == 7]
mes_julio
```

```python
#Filtramos las ventas donde venden accesorios
ventas_accesorios = df[df['Categoría'] == 'Accesorios']
ventas_accesorios
```

```python
#Filtramos precios, donde tiene que ser mayor a 100 y menor a 500
precio_300_500 = df[(df['Precio Unitario']>100) & (df['Precio Unitario']<500)]
precio_300_500
```

```python
#Agrupamos por categoria y obtenmos el total de cada cat
df.groupby('Categoría')['Total Venta'].sum()
```

```python
#Agruo por mes y obtenemos el total de todos los meses
df.groupby('Mes')['Total Venta'].sum()
```

```python
#Top 3 de prod más vendidos
df.groupby('Producto')['Total Venta'].max().head(3)
```
