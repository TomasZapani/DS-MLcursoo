```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Datos_de_Ventas_de_Ropa.csv')
df.head(4)
```

1_

```python
#Transformamor la serie fecha a tipo datetime
df['Fecha'] =  pd.to_datetime(df['Fecha'])
type(df['Fecha'][0])
```

```python
#Creamos dentro del df una columna solo para mes
df['Mes']= df['Fecha'].dt.month
```

```python
#Filtramos un mes espec√≠fico
ventas_en_enero = df[df['Mes'] == 1]
ventas_en_enero
```

```python
#Filtramos una categor√≠a 
Verano = df[df['Categor√≠a'] == 'Verano']
Verano
```

```python
#Prod que hizo mas ventas 
Prod_m√°s_vendido = df.groupby('Producto')['Total Venta'].max()
Prod_m√°s_vendido.head(1)
```

```python
#Prod que m√°s vendio
Prod_m√°s_vendido = df.groupby('Producto')['Cantidad'].max()
Prod_m√°s_vendido.head(1)
```

3_

```python
#Suma de las ventas totales por categoria
df.groupby('Categor√≠a')['Total Venta'].sum()
```

```python
#Suma de las ventas totales por mes
df.groupby('Mes')['Total Venta'].sum()
```

```python
#Ver tendencia de aumento o disminuci√≥n
Promedio_prod_vendidos = df.groupby('Mes')['Precio Unitario'].mean()#Obtenemos el promedio de precio unitarios por mes
variacion_precios = Promedio_prod_vendidos.pct_change() * 100#Esa fn pct_change() calc un mes con el anterior(por el 1ero es NaN pq no tiene con quien compararse) y calcula la variacion en terminos relativo y dps lo multplic por 100 para dejarlo en porcentaje al result

print(f'Variaci√≥n mensual:{variacion_precios}')
```
