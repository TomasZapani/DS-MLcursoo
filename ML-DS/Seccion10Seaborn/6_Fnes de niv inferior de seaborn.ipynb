```python
import seaborn as sns
%matplotlib inline
```

```python
pinguinos = sns.load_dataset('penguins')
```

Representamos todas las relaciones del dataset con la fn .PairGrid():

```python
#Esta fn nos permite generar un paird grid( forma de visualizar relaciones entre varias variables de un conjunto de datos) personalizado y nos permit mostrar la relacion entre pares de variables de pinguinos y nos ayuda a dif los datos por especie
g = sns.PairGrid(pinguinos,
                hue = 'species')
```

```python
Modif de un atrib de pair grid:
```

```python
g = sns.PairGrid(pinguinos,
                hue = 'species',
                corner=True)#Activando el corner lo q hago es los gráf que se repiten estilo:ejex=bill-ejeY=Flipper/ejex = Flipper-ejey=bill no se repitan-desaparecen, solo una vez, no nos ref a los graf q son iguales iguales estilo:ejex=bill-ejeY=bill/ejex=Flipper-ejey=Filpper esos siguen iguales
```

Agregacion de elementos para personalizar nuestro gráf:

Agrgamos gráficos de la parte inferior de la mátriz(no se gráf en los últ diagonales). Todo eso aplicando una fn .map_lower() q evita el gráf en la diagonales

```python
g = sns.PairGrid(pinguinos,
                hue = 'species',
                corner=True)
#.map_lower()
g.map_lower(sns.kdeplot,#Muestra la distribución de cada par de variables con la estmiación de densidad de kernel.
            hue=None,#Acá estamos anulando el hue de arriba('species') para q en esta capa de gráf el kdeplot se dib de un color único
            level =5,#Con level 5 estamos específicando el nro de niv de contorno para el gráf kde
            color='.2'#Define el color de los gráf kde usando un tono gris en este caso
           );
```

Añadimos gráf de dispersion encimma del kdeplot

```python
g = sns.PairGrid(pinguinos,
                hue = 'species',
                corner=True)
#.map_lower()
g.map_lower(sns.kdeplot,#Muestra la distribución de cada par de variables con la estmiación de densidad de kernel.
            hue=None,#Acá estamos anulando el hue de arriba('species') para q en esta capa de gráf el kdeplot se dib de un color único
            level =5,#Con level 5 estamos específicando el nro de niv de contorno para el gráf kde
            color='.2'#Define el color de los gráf kde usando un tono gris en este caso
           )
g.map_lower(sns.scatterplot,
           marker = '+');#Le metemos un scatterplot para el grá de dispersión y q en vez de puntos veamos el +
```

Gráficamos ahora en la parte diagonal con la fn .map_diag()

```python
g = sns.PairGrid(pinguinos,
                hue = 'species',
                corner=True)
#.map_lower()
g.map_lower(sns.kdeplot,#Muestra la distribución de cada par de variables con la estmiación de densidad de kernel.
            hue=None,#Acá estamos anulando el hue de arriba('species') para q en esta capa de gráf el kdeplot se dib de un color único
            level =5,#Con level 5 estamos específicando el nro de niv de contorno para el gráf kde
            color='.2'#Define el color de los gráf kde usando un tono gris en este caso
           )
g.map_lower(sns.scatterplot,
           marker = '+')
g.map_diag(sns.histplot,#Queremos q nuestras diasgonales tengan gráf de estilo histograma
          element = 'step',#Le metemos a nuestro histograma un estilo de pasos sin relleno
          linewidth = 0,#El ancho de la línea q sea 0 y elimine el borde del hist con esto
          kde=True);#Con el kde true añadimos una línea del kde sobre el histograma para mostrar las distrib, de forma suavizada
#Acá estamos colocando la fn map_diag para trab ahora con las diagonales vacías
```

Añadimos una leyenda

```python
g = sns.PairGrid(pinguinos,
                hue = 'species',
                corner=True)
#.map_lower()
g.map_lower(sns.kdeplot,#Muestra la distribución de cada par de variables con la estmiación de densidad de kernel.
            hue=None,#Acá estamos anulando el hue de arriba('species') para q en esta capa de gráf el kdeplot se dib de un color único
            level =5,#Con level 5 estamos específicando el nro de niv de contorno para el gráf kde
            color='.2'#Define el color de los gráf kde usando un tono gris en este caso
           )
g.map_lower(sns.scatterplot,
           marker = '+')
g.map_diag(sns.histplot,#Queremos q nuestras diasgonales tengan gráf de estilo histograma
          element = 'step',#Le metemos a nuestro histograma un estilo de pasos sin relleno
          linewidth = 0,#El ancho de la línea q sea 0 y elimine el borde del hist con esto
          kde=True)#Con el kde true añadimos una línea del kde sobre el histograma para mostrar las distrib, de forma suavizada
#Acá estamos colocando la fn map_diag para trab ahora con las diagonales vacías

g.add_legend(frameon=True);#Añadimos una leyenda, con el frameon = True le decimos q queremos q la leyenda tenga un marco
```

Añadimos una leyenda y cambiamos la posición de esta

```python
g = sns.PairGrid(pinguinos,
                hue = 'species',
                corner=True)
#.map_lower()
g.map_lower(sns.kdeplot,#Muestra la distribución de cada par de variables con la estmiación de densidad de kernel.
            hue=None,#Acá estamos anulando el hue de arriba('species') para q en esta capa de gráf el kdeplot se dib de un color único
            level =5,#Con level 5 estamos específicando el nro de niv de contorno para el gráf kde
            color='.2'#Define el color de los gráf kde usando un tono gris en este caso
           )
g.map_lower(sns.scatterplot,
           marker = '+')
g.map_diag(sns.histplot,#Queremos q nuestras diasgonales tengan gráf de estilo histograma
          element = 'step',#Le metemos a nuestro histograma un estilo de pasos sin relleno
          linewidth = 0,#El ancho de la línea q sea 0 y elimine el borde del hist con esto
          kde=True)#Con el kde true añadimos una línea del kde sobre el histograma para mostrar las distrib, de forma suavizada
#Acá estamos colocando la fn map_diag para trab ahora con las diagonales vacías

g.add_legend(frameon=True);#Añadimos una leyenda, con el frameon = True le decimos q queremos q la leyenda tenga un marco
g.legend.set_bbox_to_anchor((.61,.6));#Acá con esta fn conf la posición de la leyenda, en este caso le metí la posición q puso la persona en el curso pero es cuestion de ir probando(el 1er num le corresp a x y 2do a y)
```
