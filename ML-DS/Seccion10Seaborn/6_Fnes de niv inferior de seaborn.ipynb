```python
import seaborn as sns
%matplotlib inline
```

```python
pinguinos = sns.load_dataset('penguins')
```

Representamos todas las relaciones del dataset con la fn .PairGrid():

```python
#Esta fn nos permite generar un paird grid personalizado y nos permit mostrar la relacion entre pares de variables de pinguinos y nos ayuda a dif los datos por especie
g = sns.PairGrid(pinguinos,
                hue = 'species')
```

```python
g = sns.PairGrid(pinguinos,
                hue = 'species',
                corner=True)#Activando el corner lo q hago es los gráf que se repiten estilo:ejex=bill-ejeY=Flipper/ejex = Flipper-ejey=bill no se repitan-desaparecen, solo una vez, no nos ref a los graf q son iguales iguales estilo:ejex=bill-ejeY=Flipper/ejex=Flipper = Flipper-ejey=Filpper esos siguen iguales
```

Agregacion de elementos para personalizar nuestro gráf:

Agrgamos gráficos de la parte inferior de la mátriz(no se gráf en los últ diagonales). Todo eso aplicando una fn .map_lower()

```python
g = sns.PairGrid(pinguinos,
                hue = 'species',
                corner=True)
#.map_lower()
g.map_lower(sns.kdeplot,#Muestra la distribución de cada par de variables con densidad de kernel.
            hue=None,#Acá estamos anulando el hue de arriba para q en esta capa de gráf el kdeplot se dib de un color único
            level =5,#Con level 5 estamos específicando el nro de niv de contorno para el gráf kde
            color='.2'#Define el color de los gráf kde 
           );
```
