```python
import seaborn as sns
%matplotlib inline
```

```python
pinguinos = sns.load_dataset('penguins')
```

Representamos todas las relaciones del dataset con la fn .PairGrid():

```python
#Esta fn nos permite generar un paird grid( forma de visualizar relaciones entre varias variables de un conjunto de datos) personalizado y nos permit mostrar la relacion entre pares de variables de pinguinos y nos ayuda a dif los datos por especie
g = sns.PairGrid(pinguinos,
                hue = 'species')
```

```python
Modif de un atrib de pair grid:
```

```python
g = sns.PairGrid(pinguinos,
                hue = 'species',
                corner=True)#Activando el corner lo q hago es los grf que se repiten estilo:ejex=bill-ejeY=Flipper/ejex = Flipper-ejey=bill no se repitan-desaparecen, solo una vez, no nos ref a los graf q son iguales iguales estilo:ejex=bill-ejeY=bill/ejex=Flipper-ejey=Filpper esos siguen iguales
```

Agregacion de elementos para personalizar nuestro grf:

Agrgamos grficos de la parte inferior de la mtriz(no se grf en los lt diagonales). Todo eso aplicando una fn .map_lower() q evita el grf en la diagonales

```python
g = sns.PairGrid(pinguinos,
                hue = 'species',
                corner=True)
#.map_lower()
g.map_lower(sns.kdeplot,#Muestra la distribucin de cada par de variables con la estmiacin de densidad de kernel.
            hue=None,#Ac estamos anulando el hue de arriba('species') para q en esta capa de grf el kdeplot se dib de un color nico
            level =5,#Con level 5 estamos especficando el nro de niv de contorno para el grf kde
            color='.2'#Define el color de los grf kde usando un tono gris en este caso
           );
```

Aadimos grf de dispersion encimma del kdeplot

```python
g = sns.PairGrid(pinguinos,
                hue = 'species',
                corner=True)
#.map_lower()
g.map_lower(sns.kdeplot,#Muestra la distribucin de cada par de variables con la estmiacin de densidad de kernel.
            hue=None,#Ac estamos anulando el hue de arriba('species') para q en esta capa de grf el kdeplot se dib de un color nico
            level =5,#Con level 5 estamos especficando el nro de niv de contorno para el grf kde
            color='.2'#Define el color de los grf kde usando un tono gris en este caso
           )
g.map_lower(sns.scatterplot,
           marker = '+');#Le metemos un scatterplot para el gr de dispersin y q en vez de puntos veamos el +
```

Grficamos ahora en la parte diagonal con la fn .map_diag()

```python
g = sns.PairGrid(pinguinos,
                hue = 'species',
                corner=True)
#.map_lower()
g.map_lower(sns.kdeplot,#Muestra la distribucin de cada par de variables con la estmiacin de densidad de kernel.
            hue=None,#Ac estamos anulando el hue de arriba('species') para q en esta capa de grf el kdeplot se dib de un color nico
            level =5,#Con level 5 estamos especficando el nro de niv de contorno para el grf kde
            color='.2'#Define el color de los grf kde usando un tono gris en este caso
           )
g.map_lower(sns.scatterplot,
           marker = '+')
g.map_diag(sns.histplot,#Queremos q nuestras diasgonales tengan grf de estilo histograma
          element = 'step',#Le metemos a nuestro histograma un estilo de pasos sin relleno
          linewidth = 0,#El ancho de la lnea q sea 0 y elimine el borde del hist con esto
          kde=True);#Con el kde true aadimos una lnea del kde sobre el histograma para mostrar las distrib, de forma suavizada
#Ac estamos colocando la fn map_diag para trab ahora con las diagonales vacas
```

Aadimos una leyenda

```python
g = sns.PairGrid(pinguinos,
                hue = 'species',
                corner=True)
#.map_lower()
g.map_lower(sns.kdeplot,#Muestra la distribucin de cada par de variables con la estmiacin de densidad de kernel.
            hue=None,#Ac estamos anulando el hue de arriba('species') para q en esta capa de grf el kdeplot se dib de un color nico
            level =5,#Con level 5 estamos especficando el nro de niv de contorno para el grf kde
            color='.2'#Define el color de los grf kde usando un tono gris en este caso
           )
g.map_lower(sns.scatterplot,
           marker = '+')
g.map_diag(sns.histplot,#Queremos q nuestras diasgonales tengan grf de estilo histograma
          element = 'step',#Le metemos a nuestro histograma un estilo de pasos sin relleno
          linewidth = 0,#El ancho de la lnea q sea 0 y elimine el borde del hist con esto
          kde=True)#Con el kde true aadimos una lnea del kde sobre el histograma para mostrar las distrib, de forma suavizada
#Ac estamos colocando la fn map_diag para trab ahora con las diagonales vacas

g.add_legend(frameon=True);#Aadimos una leyenda, con el frameon = True le decimos q queremos q la leyenda tenga un marco
```
