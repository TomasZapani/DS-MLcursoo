```python
pinguinos = sns.load_dataset('penguins')
```

Representamos todas las relaciones del dataset con la fn .PairGrid():

```python
#Esta fn nos permite generar un paird grid personalizado y nos permit mostrar la relacion entre pares de variables de pinguinos y nos ayuda a dif los datos por especie
g = sns.PairGrid(pinguinos,
                hue = 'species')
```

Agregacion de elementos para personalizar nuestro grf:

Agrgamos grficos de la parte inferior de la mtriz(no se grf en los lt diagonales). Todo eso aplicando una fn .map_lower()

```python
g = sns.PairGrid(pinguinos,
                hue = 'species',
                corner=True)
#.map_lower()
g.map_lower(sns.kdeplot,#Muestra la distribucin de cada par de variables con densidad de kernel.
            hue=None,#Ac estamos anulando el hue de arriba para q en esta capa de grf el kdeplot se dib de un color nico
            level =5,#Con level 5 estamos especficando el nro de niv de contorno para el grf kde
            color='.2'#Define el color de los grf kde 
           );
```
