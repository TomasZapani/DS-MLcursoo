```python
import seaborn as sns
%matplotlib inline
```

```python
#Cargamos los datos
propinas = sns.load_dataset('tips')
propinas;
```

stripplot:gráfico de dispersion pero una variable es enfocada a lo categórico como día acá

```python
sns.catplot(
    data = propinas,
    kind = 'strip',#Será un gráf de tipo stipplot
    x = 'day',#en el eje x irán los días, esto es una variable categórica no numérica pq almacena únicamnete los días, osea... strings
    y='total_bill',
    hue = 'smoker'
);
```

Swarm:es una variacion del strip, únicamente que en este los puntos no se solapan-superponen pq se los distrib de forma perpendicular para que cada punto sea visible y que facilite la identif de la densidad de la estructura de los datos 

```python
sns.catplot(
    data = propinas,
    kind = 'swarm',#Será un gráf de tipo swarmplot
    x = 'day',
    y='total_bill',
    hue = 'smoker'
);
```

Gráf boxplot:Básicamente es un gráf que muestra como se distrib los datos, ayuda a ver q tan dispersos están,donde está la mitad de los valores y si hay val extremos

```python
sns.catplot(
    data = propinas,
    kind = 'box',#Será un gráf de tipo boxplot
    x = 'day',
    y='total_bill',
    hue = 'smoker'
);
```

Graf violinplot:muestra la densidad de los datos en dif valores, básicamente muestra cuantos datos hay en cada valor

```python
sns.catplot(
    data = propinas,
    kind = 'violin',#Será un gráf de tipo violinplot
    x = 'day',
    y='total_bill',
    hue = 'smoker'
);
```

Gráf pointplot:muestra las estimaciones puntuales con sus intervalos usando marcadores de puntos

```python
sns.catplot(
    data = propinas,
    kind = 'point',#Gráf de pointplot
    x = 'day',
    y='total_bill',
    hue = 'smoker'
);
```

Gráf barplot:Básicaente ete gráf sirve para representar la relacion entre var numérica y cat, cada barra representa un categoría y la altura representa la variable numérica de esta categoria. La línea que sobresale muestra que tanta incertidumbre hay en la alt de nuestro gráf: linea ++ baja  o ++ alta signif que podría ser mas baja o mas alta aun incertidumbre y linea --baja o linea -- alta signif rango normal y no hay tanta variación  

```python
sns.catplot(
    data = propinas,
    kind = 'bar',#Gráf barplot
    x = 'day',
    y='total_bill',
    hue = 'smoker'
);
```
