4_ jointplot:utilizamos las variables precio y quilate

```python
df_sample = diamonds.sample(n=2000, random_state=42) # Selecciona aleatoriamente 2000 filas del DataFrame 'diamonds' para crear una muestra reproducible.
g = sns.jointplot(df_sample,
            x ='carat',#Tamaño de las alas
             y='price',#Tamaño de los picos#leyenda
             hue = 'cut',
             alpha = 1);
g.set_axis_labels('Quilate','Precio')#Colocamos los nombre para ambos ejes
legend = g.ax_joint.get_legend()#Accedo a la leyenda atravez del eje y....
legend.set_title("Calidad")#Coloco el título(asi se accede y se hace con jointplot)
```

Inisight:En resumen, el gráfico confirma que el quilataje es un factor determinante en el precio (relación creciente), pero también evidencia que hay otros factores (como el corte) que influyen en la dispersión de los precios
