```python
import seaborn as sns
```

```python
pinguinos = sns.load_dataset('penguins')
pinguinos
```

```python
Crearemos un gráfico estilo histograma sobre el largo de los brazos de los pinguinos y los separaremos por especie
```

```python
sns.displot(data = pinguinos,#Le pasamos la data q queremos gráf
            kind= 'hist',#displot haceme un gráf de tipo histograma
           x = 'flipper_length_mm',
           hue = 'species',#separamos por especie
           multiple = 'stack');#Acá lo q hace el parám mutiple con su valor stack es superponer las distribuciones de cada especie, las apila una encima de otra en el histograma.
```

```python
Lo hacemos con un gráf de bajo niv(con histplot)
```

```python
sns.histplot(data = pinguinos,#Le pasamos la data q queremos gráf
           x = 'flipper_length_mm',
           hue = 'species',#separamos por especie
           multiple = 'stack');#Acá lo q hace el parám mutiple con su valor stack es superponer las distribuciones de cada especie, las apila una encima de otra en el histograma.
```

```python
Demostracion de como la fn madre de displot maneja varios axes:
```

```python
sns.displot(data = pinguinos,#Le pasamos la data q queremos gráf
            kind= 'hist',#displot haceme un gráf de tipo histograma
            col = 'species',
           x = 'flipper_length_mm',
           hue = 'species',#separamos por especie
           );
```

```python
sns.relplot(data=pinguinos,
           x = 'flipper_length_mm',
           y = 'bill_length_mm',
           col = 'sex');
```

Colocamos nombres a los ejes

```python
g = sns.relplot(data=pinguinos,
           x = 'flipper_length_mm',
           y = 'bill_length_mm',
           col = 'sex')


g.set_axis_labels('Largo aletas (en mm)','Largo pico (en mm)');
```
