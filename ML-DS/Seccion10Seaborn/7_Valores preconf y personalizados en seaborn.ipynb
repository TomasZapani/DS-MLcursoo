```python
import seaborn as sns
import matplotlib as plt
%matplotlib inline
```

```python
pinguinos = sns.load_dataset('penguins')
```

```python
sns.relplot(data=pinguinos,
           x = 'bill_length_mm',
           y = 'bill_depth_mm',
           hue = 'body_mass_g');
```

relplot es genial para mostrarnos un grf hecho por l con sus cositas y en base a lo que el sabe, etc etc. Pero ahora bien... si nuetsra idea es mejorarlo-pulirlo, seaborn nos ofrece varios niv de personlaizacin y otros recursos

Vamos a hacer un ejemplo en el cual le cambiamos los parmetros de lo q nos genera seaborn para hacerlo mas personalizado

```python
sns.set_theme(style='dark')#Le damos un estilo dark


sns.relplot(data=pinguinos,
           x = 'bill_length_mm',
           y = 'bill_depth_mm',
           hue = 'body_mass_g',
           palette = 'crest',#Paleta de colores
           marker = 'x',#Q los marcadores en vez de circulos sean x
           s=100);#Q tenga un tamao de 100 cada x
```

```python
g = sns.relplot(data=pinguinos,
           x = 'bill_length_mm',
           y = 'bill_depth_mm',
           hue = 'body_mass_g',
           palette = 'crest',#Paleta de colores
           marker = 'x',#Q los marcadores en vez de circulos sean x
           s=100)#Q tenga un tamao de 100 cada x

#Cambiamos el nombre de los ejes y su posicin: x=Largo del pico, y = Grosor del pico y labelpad = 10 es la dist q debe mantener el nombre con respecto a los nros
g.set_axis_labels("Largo del pico (en mm)",
                 "Grosor del pico (en mm)",
                 labelpad=10)


;
```

```python
g = sns.relplot(data=pinguinos,
           x = 'bill_length_mm',
           y = 'bill_depth_mm',
           hue = 'body_mass_g',
           palette = 'crest',#Paleta de colores
           marker = 'x',#Q los marcadores en vez de circulos sean x
           s=100)#Q tenga un tamao de 100 cada x

#Cambiamos el nombre de los ejes y su posicin: x=Largo del pico, y = Grosor del pico y labelpad = 10 es la dist q debe mantener el nombre con respecto a los nros
g.set_axis_labels("Largo del pico (en mm)",
                 "Grosor del pico (en mm)",
                 labelpad=10)

#Cambiamos el ttulos de la leyenda y su posicin
g.legend.set_title('Masa\ncorporal (en gr)')
g.legend.set_bbox_to_anchor((1,0.5))#Ajust la posicin de la leyenda(x = 1-y=0.5)
;
```

Bsicamente lo q venimos haciendo es tomar elementos q le corresp al grf g y vamos variando y personalizandolo, aunque sea un grf de alto niv

Modif las pulgadas del ancho y el alto de la proporcion de la figura-pared

```python
g = sns.relplot(data=pinguinos,
           x = 'bill_length_mm',
           y = 'bill_depth_mm',
           hue = 'body_mass_g',
           palette = 'crest',#Paleta de colores
           marker = 'x',#Q los marcadores en vez de circulos sean x
           s=100)#Q tenga un tamao de 100 cada x

#Cambiamos el nombre de los ejes y su posicin: x=Largo del pico, y = Grosor del pico y labelpad = 10 es la dist q debe mantener el nombre con respecto a los nros
g.set_axis_labels("Largo del pico (en mm)",
                 "Grosor del pico (en mm)",
                 labelpad=10)

#Cambiamos el ttulos de la leyenda y su posicin
g.legend.set_title('Masa\ncorporal (en gr)')
g.legend.set_bbox_to_anchor((1,0.5))#Ajust la posicin de la leyenda(x = 1-y=0.5)

#Establecemos las pulgadas de nuestro grf
g.figure.set_size_inches(6,4.5)
;
```

Ampliamos el margen alrededor del grf para q los datos no se vean tan pegados a los bordes

```python
g = sns.relplot(data=pinguinos,
           x = 'bill_length_mm',
           y = 'bill_depth_mm',
           hue = 'body_mass_g',
           palette = 'crest',#Paleta de colores 
           marker = 'x',#Q los marcadores en vez de circulos sean x
           s=100)#Q tenga un tamao de 100 cada x

#Cambiamos el nombre de los ejes y su posicin: x=Largo del pico, y = Grosor del pico y labelpad = 10 es la dist q debe mantener el nombre con respecto a los nros
g.set_axis_labels("Largo del pico (en mm)",
                 "Grosor del pico (en mm)",
                 labelpad=10)

#Cambiamos el ttulos de la leyenda y su posicin
g.legend.set_title('Masa\ncorporal (en gr)')
g.legend.set_bbox_to_anchor((1,0.5))#Ajust la posicin de la leyenda(x = 1-y=0.5)

#Establecemos las pulgadas de nuestro grf
g.figure.set_size_inches(6,4.5)

#Ampliamos los margenes de nuestro grf
g.ax.margins(.15)
;
```
