```python
import seaborn as sns
```

Cargamos un dataset preedefinido en seaborn

```python
pinguinos = sns.load_dataset('penguins')#dataset penguins
pinguinos.head()
```

```python
sns.displot(pinguinos,
           x = 'body_mass_g')#Gráf para ver como se div la masa corporal de nuestros pinguinos en dist agrup
```

```python
Lo de arriba era un histograma
```

Añadimos un gráf kde: este gráf lo q hace básicamente es mostrarnos donde se concentran mas datos y ver como están ditribuidos los mismos, de forma suave representada con curvas

```python
sns.displot(pinguinos,
           x = 'body_mass_g',
           kde='True')#Activamos kde con True
```

```python
Añadimos las columnas para dif por islas la masa corporal
```

```python
sns.displot(pinguinos,
           x = 'body_mass_g',
           kde='True',
           col = 'island')#metemos el parám col y le definimos que lo queremos separar por islas
```

Creamos un gráf para reprsenta el largo de las aletas por mm, el gráfico será de tipo kde y vamos activar el complemento rugplot.
rugplot es un complemento no un gráf q sirve para ver la densidad-distribucion de los datos en un eje mediante flecos

```python
sns.displot(pinguinos,
           x = 'flipper_length_mm',
           kind = 'kde',#q el gráf se base en kde
           rug = True)#Activamos el rugplot
```

Ahora vamos a gráf el ecdfplot: es un gráf el cpual nos muestra como va aumentando nuestro gráf en base a los valores acumulados por cada nro....:

```python
sns.displot(
    pinguinos,
    x = 'flipper_length_mm',
    kind = 'ecdf'#Usamos el ecdfplot gracias al parám kind
)
```
