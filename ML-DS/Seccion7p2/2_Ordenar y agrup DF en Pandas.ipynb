```python
import pandas as pd
```

```python
df = pd.read_csv('Top-Películas.csv')
df
```

```python
#Ordenar datos del df, en este caso específico por rating
df_ordenado = df.sort_values(by = 'rating')#Lo va a ordenar de menor a mayor gracias a la fn sort...
df_ordenado.head(40)
```

```python
#Ordenado del df pero ahora de mayor a menor...
df_ordenado = df.sort_values('rating',ascending=False)#Esto de ordenar al revez lo logramos pq desactivamos el ascending y le metimos False
df_ordenado
```

```python
#Ordenamiento en funcion de varias columnas
df_ordenado = df.sort_values(by=['rating','recaudación(M)'],ascending = False)#Le metemos otra columna colocando el nombre de la misma en la lista
df_ordenado
```

```python
#Agrupación de los géneros de las películas y obtener el promedio de rating de cada uno de esos géneros para el orden
df_agrupado = df.groupby('género')['rating'].mean()#Agrupamos gracias a groupby ahí le ponemos el name del género, dps le decimos que queremos ordenar ese grupo por el rating y aplicarle la fn mean q es para obtener el promedio de rating de cada género
df_agrupado#Impresión para obtener la agrup de los géneros, ordenados por el promedio del rating de cada uno
```

```python
#Agrupacion de las películas por el año y ordenar los años por la recaudación de cada uno
df_agrupado = df.groupby('año')['recaudación(M)'].sum()#Agrupamos por año y ordenamos por la recaudación total(suma) de cada año  
df_agrupado.sort_values(ascending = False).head(10)#Aplicamos la fn sort... para ordenar esto de mayor a menor y que nos muestre los 10 1eros
```
