```python
import matplotlib.pyplot as plt
%matplotlib inline
```

Creación de los multiples gráficos...:

```python
#Acá creo dos col con dos filas cada col
fig, axs = plt.subplots(ncols=2,nrows=2)#Acá estamos creando 4 gráf en total, metiendoles dos columnas y dos filas todo esto almacenanadolo en dos variables
```

```python
#Acá creo dos cols y que cada columna tenga una úncia fila
fig, axs = plt.subplots(ncols=2,nrows=1)
```

Indexar cada uno de estos dos axes para desplegar un gráf distintos en cada uno de ellos

```python
fig, axs = plt.subplots(ncols=2,nrows=1)

#Estamos guardando los dos subgráficos en variables distintas
ax1 = axs[0]
ax2 = axs[1]
;
```

Dibujamos dos gráficos con los mismo datos en los dos axes

```python
#Ahora realizamos dos gráficos iguales(uno de línea y otro scatter) con los mismo datos
fig, axs = plt.subplots(ncols=2,nrows=1)

#Estamos guardando los dos subgráficos en variables distintas
ax1 = axs[0]
ax2 = axs[1]

ax1.plot([1,2,3,4],[10,32,34,4])
ax2.scatter([1,2,3,4],[10,32,34,4])

;
```

```python
#Ahora realizamos dos gráficos iguales(uno de línea y otro scatter) con los mismo datos
fig, axs = plt.subplots(ncols=2,nrows=1)

#Estamos guardando los dos subgráficos en variables distintas
ax1 = axs[0]
ax2 = axs[1]

ax1.plot([1,2,3,4],[10,32,34,4])
ax2.scatter([1,2,3,4],[10,32,34,4])

#Aplicamos título gracias al método .set_title del objeto Axes q es una instancia de la clase matplotlib.axes.Axes de la libreria matplotlib, asi se titula cuando hay mas de un gráf 
ax1.set_title('Gráfico de línea')
ax2.set_title('Gráfico de Scatter')
;
```
