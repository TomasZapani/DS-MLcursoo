```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = [1,2,3,4,5,4,3,5,3]
y = [1,2,4,6,4,3,2,1,5]

plt.grid()
plt.plot(x,y)#plot se encarga de tomar toda la info que le pasemos y se encarga de gráf como línea o marcadores
```

Usos de nomenclaturas para format(está mas específico en la notebook que de nomenclaturas que dejo el porfe para esta seccion)

```python
plt.plot(x,y,'ro-')#Acá le metemos el color rojo, unos circulos por la o y la lineas por la -
```

```python
plt.plot(x,y,'b*:')#Acá le metemos de color blue, en vez de circulo que sea tipo estrella y en vez de linea recta que sean puntos
```

Conf el val máximo y mínimo del gráfico

```python
plt.plot(x,y,'b*:')
plt.axis([0,8,0,8])#Este método axis sirve para personalizar los ejes, en este caso le estamos conf el min y max del eje x e y.Un mínimo de 0 y un máx de 8 para cada eje, los primeros dos son para el eje x y los otros dos para el y 
plt.grid()
```
