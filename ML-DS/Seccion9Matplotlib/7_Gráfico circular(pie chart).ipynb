```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

Es un gr√°fico bastante conocido y facil de realizar

Creamos el df para el ej

```python
df = pd.DataFrame({
    'etiquetas':['Manzanas','Bananas','Naranjas','Uvas'],
    'cant':[40,32,20,14]
})

df
```

```python
Graficamos la torta de este una serie de este df
```

```python
plt.pie(df['cant']);#El m√©todo para graf el pastel es .pie()
```

Manejo de par√°metro con el gr√°f circular

```python
#Mostrar las etiquetas en el pastel para dif, con el param label
plt.pie(df['cant'],labels=df['etiquetas']);#Le pasamos a labels que queremos mostrar, dandole la serie de etiquetas del df
```

```python
#Vamos a ver los porcentajes del pastel con el param autopct
plt.pie(df['cant'],labels=df['etiquetas'],autopct='%1.1f%%');#Le pasamos a autopct la forma de como queremos que se vea el porcent, hay que poner los dos %%, dps en el medio se pone el 1.1%, donde el el primer 1 es el numero entero y segundo 1 es para que vea que quiero solo un decimal dps del punto,la f obvio es de float y el % que est√° adentro indica que queremos que estos porcentajes se vean con su simbolo 
```

```python
#Manip de los colores del gr√°f con el param colors
plt.pie(df['cant'],labels=df['etiquetas'], autopct='%1.1f%%',colors = ['b','y','g','orange']);#Le pasamos los colores en formato de lista, algunos con sus iniciales y otros escritos entero
```

```python
#Agregamos sombras a nuestro gr√°f con el par√°m shadow activandolo con True
plt.pie(df['cant'],labels=df['etiquetas'], autopct='%1.1f%%',colors = ['b','y','g','orange'],shadow=True);
```

```python
#Separamos una porcion de gr√°f para destacarla con el par√°m explode
plt.pie(df['cant'],labels=df['etiquetas'], autopct='%1.1f%%',colors = ['b','y','g','orange'],shadow=True,explode=(0,0,0,0.5));#Le tenemos que dar al par√°m la distancia de cada porcion y le dimo 0 al primero,0 al 2do, 0 al 3ero y 0.5 al cuarto
```

```python
#Le metemos un t√≠tulo al gr√°fico con la fn .title()
plt.pie(df['cant'],labels=df['etiquetas'], autopct='%1.1f%%',colors = ['b','y','g','orange'],shadow=True,explode=(0,0,0,0.5));
plt.title('Frutas');#Le pasamos el nombre como argumento
```
