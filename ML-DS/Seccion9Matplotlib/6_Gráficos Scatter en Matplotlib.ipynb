```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

Gráfico scatter(desparramar en ingles):toman los registros individuales que hay en nuestros datos y los desparrama como puntos en una cuadricula

```python
#Creacion de nuestros datos
x = np.linspace(0,10,30)#Este método lo que hace es crear un conj de dato que va desde 0 y 10 y que cree dentro de ese limite 30 números con una distancia similar entre ellos
y = np.sin(x)#Acá y va tener la misma cant de numeros que x, pq estamos sacando el seno de los números de x
```

```python
#A ver como quedo nuestro gráfico representado en forma de línea....
plt.plot(x,y);
```

```python
#Representamos nuestro gráfico ahora en forma de puntos-criculos, gracias al 3er arg 'o' que representa el tipo de marcador de puntos que usa plot para representar nuestros puntos
plt.plot(x,y,'o');
```

Creamos un codigo para crear y mostrra un gráf con dist simbolos y representarlos en dif puntos aleatorios

```python
aleatorios = np.random.RandomState(0)#La fn lo que hace es ser un contenenedor que genera numeros aleatorios pero siempre los que genera son siempre iguales 
marcadores = ['o','x','+','v','s','d']#Esto son los tipos de marcadores que hay
```

```python
#Bucle para crear representaciones en scatter dentro del plot tantas veces segun cuanto marcadores haya
for marcador in marcadores:
    plt.plot(aleatorios.rand(5), aleatorios.rand(5), marcador)#Generamos el graf pasandole al eje x y al eje y para q generen 5 puntos en total para cada marcador tanto para x e y(usando el generador que ya habiamos creado arriba con RandomState) y le pasamos el argumento marcador para que grafique esos marcadores que recorrió
```

Creaciónde gráficos probando varios parámetros de personalizacion

```python
#Graficamos lo mismo que el 1ero pero le metemos un formatos distintio, lineas punteadas,circulos y color azul
plt.plot(x,y,':ob');
```

```python
#Acá le aplicamos lineas sólidas, pentágonos y de color amarillo
plt.plot(x,y,'-py');
```

```python
#Acá hacemos lo mismo que arriba pero el valor color lo definimos con su respectivo parámetro
plt.plot(x,y,'-p', color='green');
```

```python
#Acá hacemos lo mismo solo que aumentamos el tamaño del marcador para que esté más grueso
plt.plot(x,y,'-py', color='red',markersize=15);
```

```python
#Acá le metemos un linewidth para agrandar el ancho de la línea
plt.plot(x,y,'-p', color='red',markersize=6,linewidth=4);
```

```python
#Acá le metimos color a los marcadores del gráfico
plt.plot(x,y,'-p', color='red',markersize=6,linewidth=4,markerfacecolor='green');
```

Uso del método Scatter

```python
#Método específico para gráficos scatter, se lo puede hacer con plot, pero es más personalizable y controlado con scatter
plt.scatter(x,y);
```

```python
#Usamos ahora el método random para generar 100 puntos
aleatorio = np.random.RandomState(0)
plt.scatter(aleatorio.randn(100),aleatorio.randn(100));
```

```python
#Es lo mismo que arriba pero le metimos que genere 100 colores randoms
aleatorio = np.random.RandomState(0)
plt.scatter(aleatorio.randn(100),aleatorio.randn(100),c=aleatorio.randn(100));
```

```python
#Ahora le metimos el parámetro s que es size y le multplicamos el tamaño de los puntos
aleatorio = np.random.RandomState(0)
plt.scatter(aleatorio.randn(100),aleatorio.randn(100),c=aleatorio.randn(100),s=aleatorio.randn(100) *1000);
```

```python
#Ahora le metimos el parámetro alpha que es para manejar la opacidad de los circulos
aleatorio = np.random.RandomState(0)
plt.scatter(aleatorio.randn(100),aleatorio.randn(100),c=aleatorio.randn(100),s=aleatorio.randn(100) *1000, alpha = 0.3);
```

```python
#Ahora le metimos el parámetro cmap que lo que hace es mapear el gráfico y en base a los valores numericos que haya va asignando un color en escala y hay que pasarle valores, en este caso le pasamos viridis para que eliga como asignar pero hay mas valores
aleatorio = np.random.RandomState(0)
plt.scatter(aleatorio.randn(100),aleatorio.randn(100),c=aleatorio.randn(100),s=aleatorio.randn(100) *1000, alpha = 0.3,cmap='viridis');
```

```python
#Ahora le metimos un método llamado colorbar() para q nos ayude a ver ,los colores que hay en gráfico
aleatorio = np.random.RandomState(0)
plt.scatter(aleatorio.randn(100),aleatorio.randn(100),c=aleatorio.randn(100),s=aleatorio.randn(100) *1000, alpha = 0.3,cmap='viridis');
plt.colorbar()
```
