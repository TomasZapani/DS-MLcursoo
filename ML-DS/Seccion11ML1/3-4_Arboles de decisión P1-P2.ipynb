```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.preprocessing import LabelEncoder# Convierte etiquetas categóricas (strings) a valores numéricos para que puedan ser procesados por modelos de ML.
from sklearn.model_selection import train_test_split# Divide el dataset en subconjuntos de entrenamiento y prueba, lo que ayuda a evaluar el rendimiento del modelo.
from sklearn.tree import DecisionTreeClassifier,plot_tree# 'plot_tree' permite visualizar el árbol generadoy decisiontreeclasifier contiene el modelo del árbol de decision
```

Cargamos un dataset q cada fila corresp a la medicion de una flor de iris y entrenaremos un modelo para q en base a las características de la flor nos diga a q especie pertenece

```python
df = sns.load_dataset('iris')
df
```

```python
df.species.unique()#Cuantos tipos de especies tiene iris?
```

Eliminamos la parte de especies-sus registros y dejamos el resto de las columnas con sus registros

```python
X = df.drop('species',axis = 1)
X
```

Tranformamos las especies q hay en la columna 'species' en números

```python
le = LabelEncoder()
especies = le.fit_transform(df['species'])
especies
```

```python
y = especies
y
```

```python
Div el conj de datos en entrenamiento y prueba
```

```python
X_entrena,X_prueba,y_entrena,y_prueba = train_test_split(X,y,train_size = 0.8,random_state = 42)
```

```python
Creamos el árbol
```

```python
arbol = DecisionTreeClassifier()
```

```python
Lo entrenamos
```

```python
arbol.fit(X_entrena,y_entrena)
```

```python
lo gráf al árbol
```

```python
plot_tree( decision_tree = arbol)
plt.show()
```

Piensa en este árbol como un **cuestionario** paso a paso para clasificar una flor en una de tres clases (por ejemplo, setosa, versicolor o virginica):

1. **Pregunta principal (nodo raíz):**  
   - “¿El ancho del pétalo es menor o igual a 0.8?”  
   - Si **sí**, todas esas flores son *setosa*. (Porque tienen pétalos muy angostos.)

2. **Si no (rama derecha):**  
   - Se hace otra pregunta: “¿La longitud del pétalo es menor o igual a 4.75?”  
   - Dependiendo de la respuesta, sigue preguntando cosas como “¿El ancho del pétalo es menor que 1.75?” o “¿La longitud del pétalo es menor que 4.85?”  
   - Cada pregunta separa mejor las flores de la clase *versicolor* o *virginica*.

En cada “recuadro” del árbol ves:
- **samples**: cuántas flores han llegado hasta ese punto.
- **value**: cuántas de esas flores pertenecen a cada clase.
- **class**: la clase mayoritaria (o la predicción final si ya no hay más preguntas).

El objetivo es que, con preguntas simples de “sí o no” sobre las características (ancho y largo de pétalo, etc.), el árbol vaya filtrando hasta llegar a una decisión final de la clase de flor.
