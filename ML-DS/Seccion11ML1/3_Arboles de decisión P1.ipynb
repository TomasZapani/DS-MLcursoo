```python
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder# Convierte etiquetas categóricas (strings) a valores numéricos para que puedan ser procesados por modelos de ML.
from sklearn.model_selection import train_test_split# Divide el dataset en subconjuntos de entrenamiento y prueba, lo que ayuda a evaluar el rendimiento del modelo.
from sklearn.tree import DecisionTreeClassifier,plot_tree# 'plot_tree' permite visualizar el árbol generado.
```

```python
df = sns.load_dataset('iris')
df
```

```python
df.species.unique()#Cuantos tipos de especies tiene iris?
```

Eliminamos la parte de especies-sus registros y dejamos el resto de las columnas con sus registros

```python
X = df.drop('species',axis = 1)
X
```

Tranformamos las especies q hay en la columna 'species' en números

```python
le = LabelEncoder()
especies = le.fit_transform(df['species'])
especies
```

```python
y = especies
y
```

```python
Div el conj de datos en entrenamiento y prueba
```

```python
X_entrena,X_prueba,y_entrena,y_prueba = train_test_split(X,y,train_size = 0.8,random_state = 42)
```

```python
Creamos el árbol
```

```python
arbol = DecisionTreeClassifier()
```

```python
Lo entrenamos
```

```python
arbol.fit(X_entrena,y_entrena)
```

```python
lo gráf al árbol
```

```python
plot_tree( decision_tree = arbol)
plt.show()
```
