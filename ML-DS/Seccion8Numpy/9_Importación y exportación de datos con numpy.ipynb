Traemos un arch .csv a un array con un método de numpy

```python
array = np.genfromtxt('medallas.csv')#Esto generará un error, pq en nuestro arch principal se mezclan datos numericos, de texto y los arrays solo pueden contener datos del mismo tipo y por esto tendremos que meter mano en los parámetros del método genfromtxt() para adecuarlo a nuestras necesidades
array

#El met gen... es similar al read_csv() en pandas
```

```python
array = np.genfromtxt('medallas.csv',delimiter=',')#El delimitador es un param necesario para separar los valores
array
```

```python
array = np.genfromtxt('medallas.csv',delimiter=',',filling_values=0)#Acá con el filling values le estoy diciendo que los valores de tipo nan los rellene con 0, el tema ahora es que en mi primer columna tengo llena de 0 pq en la 1er columna hay texto:[  0.,   0.,   0.,   0.,   0.] y lo rellena automáticamnete de ceros y esto es un prob por ende lo solucionaremos...
array
```

```python
#Solución a lo del encabezado
array = np.genfromtxt('medallas.csv',delimiter=',',filling_values=0,skip_header=1)#Ahora metiendole parámetro skip_header y cambiandole el valor de 0 a 1 hacemos que se saltee el encabezado
array

#Ahora toca arregla el tema de que los números sean de tipo int y no float
```

```python
#Solucion
array = np.genfromtxt('medallas.csv',delimiter=',',filling_values=0,skip_header=1,dtype=int)#Para hacer q los num sean int solo le cambiamos el valor del parámtro dtype
array
```
