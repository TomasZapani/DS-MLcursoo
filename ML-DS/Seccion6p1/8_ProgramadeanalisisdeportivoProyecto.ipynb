4_Analisis de datos de medallas por país

cambiar tipo de dato de float a int 

```python
#Top 3 de países 
df['Total_medallas'] = df['Oro']+df['Plata']+df['Bronce']#Creamos una nueva columna 'Total_medallas' y dentro de esta estará la sumatoria de todas las medallas, oro,plata,bronce para ver el total de todas que tiene cada país
df_ordenar = df.sort_values(by='Total_medallas',ascending = False).head(3)#Gracias a la fn .sort_values(signif ordenar por valores) que le pasamos los args como la columna donde queremos que se plasme esto, que no sea sea ascendente sino que vaya de arriba hacía abajo para agarra ese top y dps le pasamos el método .head() con el arg 3 para que tome solo los 3 primeros registros de la columna-serie
df_ordenar['Total_medallas'] = df_ordenar['Total_medallas'].astype(int)#aca estamos haciendo ref a la columna para trab con ella:df_ordenar['Total_medallas'], dps le aplicamos la fn .astype() y dps la reasignamos a la columna original:df_ordenar['Total_medallas'] = ....Todo esto con el fin de que la cant de medallas en total no se vea con decimales, ya que tiene que ser un nro entero por lógica
df_ordenar#Mostramos el top3
```
